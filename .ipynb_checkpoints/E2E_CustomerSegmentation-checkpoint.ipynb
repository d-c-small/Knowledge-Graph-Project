```python
import numpy as np
import pandas as pd
#import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
#Load the dataset
data = pd.read_csv('usercode/customer_segmentation.csv')
data
```

```python
#Let's ecplore what the data is about
data.describe()
```

```python
#The index column is budging me and is useless so I'll drop it
data.drop(columns=['index'], inplace=True)
data
```

```python
data[data.isna().any(axis=1)]
```

```python
data.dropna(inplace=True)
```

```python
print(data.head())
data.describe()
```

```python
#Let's see how many unique invoices we have and how many unique customers there are
print("Total Customers are:", (data.CustomerID.nunique()))
print("Total Purchases made in the store are:", (data.InvoiceNo.nunique()))
```

```python
#Let's group the data based on the invoices to get total price per invoice.
#For this, we add a column named "Price" which will basically be "unit price" * "quantity"
data['Price']=data['UnitPrice']*data['Quantity']
```

```python
data

#Might reorder the columns via the command: df[['col2', 'col3', 'col1']], as I am not liking this order :(
```

## Let's see which customer made how many purchases along with their latest purchase date and the total paid amount for their purchases


```python
# Step 1: Customers and their number of purchases--->Frequency

#size(): counts number of occurrences per group
#reset_index(name='counts'): Adds name to this count column, it is unnamed otherwise3
#apply(lambda x: x): prints the result as df. Otherwise we only get the type of Structure holding it. Need to know more about this

GroupByOrders=data.groupby("CustomerID", group_keys=True)['CustomerID'].size().reset_index(name='NoOfOrders').apply(lambda x: x)  #InvoiceNo
GroupByOrders
#Here is another way of doing it, replacing 'size' with "agg('count')"
#print(data.groupby("CustomerID", group_keys=True)['CustomerID'].agg('count').reset_index(name='NoOfOrders').apply(lambda x: x))
```

```python
#Monetory Value
#Step 2: Customers and their total amount spent on all purchases
GroupByAmountSpent=data.groupby("CustomerID", group_keys=True)['Price'].agg('sum').reset_index(name='TotalAmount').apply(lambda x: x)
GroupByAmountSpent
```

```python
#Recency
LatestOrdersOfCustomers=data.groupby("CustomerID", group_keys=True)['InvoiceDate'].agg('max').reset_index(name='LatestPurchase').apply(lambda x: x)
LatestOrdersOfCustomers
```

```python
rfm_data=LatestOrdersOfCustomers
rfm_data
#restaurant_review_frame.join(restaurant_ids_dataframe, on='business_id', how='left', lsuffix="_review")
```

```python
rfm_data=rfm_data.merge(GroupByOrders, on='CustomerID', how='left')
rfm_data=rfm_data.merge(GroupByAmountSpent, on='CustomerID', how='left')
rfm_data
```

```python
#Let's rename the columns as R, F & M
rfm_data.rename(columns={'LatestPurchase' : 'Recency', 'NoOfOrders' : 'Frequency' ,'TotalAmount': 'MonetaryValue'}, inplace = True )
display(rfm_data)
```

```python
rfm_data['Recency'] = rfm_data['Recency'].astype('datetime64[ns]')
```

```python
from datetime import timedelta

#Now the recency is in date format, let's convert it into days
Snapshot_Date = (rfm_data['Recency'].max()) + timedelta(days=1)

rfm_data['Recency']=rfm_data['Recency'].apply(lambda x : (Snapshot_Date-x).days)
```

```python
rfm_data.info()
```

```python
#Let's standardize the data to avoid any biases

# standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
rfm_data_s = pd.DataFrame(scaler.fit_transform(rfm_data),columns = rfm_data.columns)

# statistics of scaled data
print(pd.DataFrame(rfm_data).describe())
#df_scaled = pd.DataFrame(ss.fit_transform(df),columns = df.columns)
print(rfm_data_s.describe())
```

```python
# Plot RFM distributions
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(40,20))

plt.subplot(3, 3, 1);
sns.histplot (rfm_data_s['Recency'])
plt.ylabel('Probability', fontsize=12)
plt.xlabel('Recency', fontsize=12)

plt.subplot(3, 3, 2);
sns.histplot (rfm_data_s['Frequency'])
plt.ylabel('Probability', fontsize=12)
plt.xlabel('Frequency', fontsize=12)

plt.subplot(3, 3, 3);
sns.histplot(rfm_data_s['MonetaryValue'])
plt.ylabel('Probability', fontsize=12)
plt.xlabel('MonetaryValue', fontsize=12)

plt.show()
```

```python
x = rfm_data_s.iloc[:,1:4] # 1st argument is to select all rows and second argument states to select column number 1 till before 3
x
```

```python
x.describe()
```

```python
wcss=[]
for i in range(1,10):
  kmeans = KMeans(i, n_init=10)
  kmeans.fit(x)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)

number_clusters = range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(5, n_init=10,init = "random", max_iter = 100, random_state = 0)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
print(identified_clusters)
np.unique(identified_clusters, return_counts=True)
```

```python
data_with_clusters = rfm_data_s.copy()
data_with_clusters['Clusters'] = identified_clusters
```
