# Create a Knowledge Graph from Text

## Task 1: Import Libraries

```python
import wikipedia as wp
import re
import requests
import spacy
import spacy_transformers
from spacy import displacy
from spacy.matcher import Matcher
import networkx as nx
from pyvis.network import Network
```

## Task 2: Import the Dataset

```python
# Set the language of the response
wp.set_lang("en")

# Obtain and store the data
query = "NewYork City"
data = wp.summary(query)

# View the data
print(data)
```

## Task 3: Preprocess the Data

```python
# Convert the data to lowercase
data = data.lower()

# Regular expression to match certain punctuation marks as well as any text within the parenthesis
pattern = r"[!@#$,;&\"']|\(.*?\)"

# Remove any text that matches the pattern.
data = re.sub(pattern, "", data)

# View the data
print(data)
```

## Task 4: Remove Stopwords 

```python
def remove_stopwords(doc):
    # Converting input to text format if it is not already.
    doc=doc.text

    # Load a language model
    nlp = spacy.load('en_core_web_lg')
    
    # Get the list of stopwords from the language model 
    stopwords = nlp.Defaults.stop_words

    word_list=[]
    for word in doc.split():
        # Store words in a list if they're not a stopword
        if word.lower() not in stopwords and word != " ":    
            word_list.append(word) 

    #Join the words to form back a string 
    data=' '.join(word_list)
    
    return data
```

## Task 5: Get Entity ID 

```python
def get_entity_id(entity_label):
    url = f"https://www.wikidata.org/w/api.php?action=wbsearchentities&format=json&search={entity_label}&language=en"
    response = requests.get(url)
    data = response.json()
    if data["search"]:
        return data["search"][0]["id"]
    else:
        return None
```

## Task 6: Recognize Named Entities

```python
#Let's create a dictionary of entities so that for the same ID, we create one node in the graph
entityDict={}

#Load a language model
nlp = spacy.load('en_core_web_lg')
doc=nlp(data)
for word in doc.ents:
    word_id=get_entity_id(word)
    
    #Create a new entry in the dictionary if the entity doesn't exist already
    if word_id not in entityDict:
        entityDict[word_id] = []
    if word.text not in entityDict[word_id]:
        entityDict[word_id].append(word.text)
    
    #Let's look at which entity is in the knowledge base and which isn't
    print(word.text, '---->' ,word.label_, '---->', word_id)
    
#Display the Entities in doc
displacy.render(doc,style="ent",jupyter=True)
```

## Task 7: Resolve Coreference

```python
#Load a language model
nlp = spacy.load('en_core_web_lg')

#Add coreference resolution component in pipeline
matcher = Matcher(nlp.vocab)
nlp.add_pipe('coreferee')

#Pass the data to the language model 
doc = nlp(data)
#Print any resolved coreferences
doc._.coref_chains.print()
```

## Task 8: Combine the Resolved Text

```python
resolved_text = ""
for token in doc:
    repres = doc._.coref_chains.resolve(token)
    if repres:
        resolved_text += " " + " and ".join([t.text for t in repres])
    elif (token.text== "."):
        resolved_text +=token.text
    else:
        resolved_text += " " + token.text
doc_resolved=nlp(resolved_text)
print(doc_resolved)
```

# Lemmatization (Optional)

```python
def lemmatize_paragraph(paragraph):
    doc = nlp(paragraph)
    lemmatized_paragraph = " ".join([token.lemma_.lower() for token in doc])
    return lemmatized_paragraph
```

## Task 9: Extract Relationships

```python
def extract_relation(doc, nlp):
    # Remove any stopwords
    doc= nlp(remove_stopwords(doc))
    
    # Get all verbs from a sentence
    verbs = [token.text for token in doc if token.pos_ == "VERB"]

    # Define a pattern
    matcher = spacy.matcher.Matcher(nlp.vocab)
    match_pattern = [{'DEP':'ROOT'}, 
          {'DEP':'prep', 'OP':"?"},
          {'DEP':'agent', 'OP':"?"},
          {'POS':'ADJ', 'OP':"?"}]

    matcher.add(key="matching_rel", patterns=[match_pattern]) 
    matches = matcher(doc)

    k = len(matches) - 1
    span = doc[matches[k][1]:matches[k][2]] 
    
    # Making sure that the relationship is a verb(if there is any)
    # If there are no verbs, go with the matcher's results
    if len(verbs) == 0:
        return span.text
    # If there are multiple verbs, select the verb that is selected by the matcher
    elif span.text in verbs:
        return span.text
    # Else return the first verb of the sentence
    else:
        return verbs[0]
```

## Task 10, 11: Extract Subject and Object

```python
# A helper function for get_sub_obj
def find_value_in_dictionary(value):
    for key, val_list in entityDict.items():
        if value in val_list:
            return key  # Return the key associated with the value
    return None  # Return None if the value is not found
```

```python
def get_sub_obj(sent):
    subject = object_ = temp = ""

    for entity in sent.ents:
        entityID= find_value_in_dictionary(entity.text)
        if entityID != None and subject != "":
            subject = entityDict[entityID][0]
            continue
     
        objectID= find_value_in_dictionary(entity.text)
        if object_== "" and objectID != None:
            object_= entityDict[objectID][0]

        if(subject !="" and object_ != "" ):
            return subject, object_
    # Make sure no stop word is selected as subject/object
    sent = nlp(remove_stopwords(sent))
    # Find subject/object from tokens
    for token in sent:
        if((token.text != " " or token.text != ".") and temp == ""):
            temp = token.text

            # Subject would be the nsubj
            if subject == "" and token.dep_ == "nsubj" and token.head.pos_ == "VERB" and token.orth_ != "" and token.orth_ != ".":
                subject = token.orth_

            # dobj is the object
            elif token.dep_ == "dobj" and token.orth_ != "." and object_=="":
                object_ = token.orth_

    # If no subject is assigned, set it to the first word of the sentence
    if(subject == ""):
        subject = temp
    
    # If no object is assigned, it is most probably the last word of the sentence
    if (object_ == ""):
        if((sent[len(sent)-3:len(sent)-2]).text == "."):
            object_ = (sent[len(sent)-2:len(sent)-1]).orth_
        else:
            object_ = (sent[len(sent)-3:len(sent)-2]).orth_

    return subject, object_
```
